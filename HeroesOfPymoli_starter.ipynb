### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
total_player = purchase_data["SN"].nunique()
totalplayer_df = pd.DataFrame({"Total Players":[total_player]})
totalplayer_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_data["Item ID"].nunique()
average_price = purchase_data["Price"].mean()
purchase_number = purchase_data["Purchase ID"].count()
total_revenue = purchase_data["Price"].sum()
purchasing_df = pd.DataFrame({"Number of Unique Items":[unique_items]
                              ,"Total Revenue":[total_revenue]
                             , "Average Price":[average_price]
                             , "Number of Purchases":[purchase_number]})
purchasing_df["Average Price"] = purchasing_df["Average Price"].map("${:.2f}".format)
purchasing_df["Total Revenue"] = purchasing_df["Total Revenue"].map("${:.2f}".format)
purchasing_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
player_count = purchase_data["Gender"].value_counts()
player_percentage = player_count/total_player*100
gender_demo = pd.DataFrame({"Percentage of Players":player_percentage
                           , "Total Count":player_count})
gender_demo["Percentage of Players"] = gender_demo["Percentage of Players"].map("{:.2f}".format)

gender_demo
# PERCENTAGE : map("{0:.2f}%".format)
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_data = purchase_data [["Gender", "Price"]]
gender_group = gender_data.groupby("Gender").sum()
gender_group["Purchase Count"] = gender_data.groupby("Gender").count()
gender_group["Average Purchase Price"] = (gender_group["Price"]/gender_group["Purchase Count"]).map("${:.2f}".format)
gender_group["Total Purchase Value"] = gender_group["Price"].map("${:.2f}".format)
gender_group["Avg Purchase Total per Person"] = gender_group["Average Purchase Price"]
del gender_group["Price"]
gender_group.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
age_demo = pd.DataFrame({"Age":purchase_data["Age"]})
age_demo[" "] = pd.cut(age_demo["Age"], age_bins, labels=group_names)
age_group = age_demo.groupby(" ").sum()
age_group["Total Count"] = age_demo.groupby(" ").count()
age_group["Percentage of Players"] = age_group["Total Count"]/total_player*100
age_group["Percentage of Players"] = age_group["Percentage of Players"].map("{:.2f}".format)
age_organized = age_group[["Percentage of Players", "Total Count"]]


age_organized
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data[" "] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)
purchase_group = purchase_data.groupby(" ").sum()
purchase_group["Purchase Count"] = age_group["Total Count"]
purchase_group["Average Purcahse Price"] = purchase_group["Price"]/purchase_group["Purchase Count"]
purchase_group["Average Purcahse Price"] = purchase_group["Average Purcahse Price"].map("${:.2f}".format)
purchase_group["Average Purcahse Total per Person"] = purchase_group["Average Purcahse Price"]
purchase_group["Price"] = purchase_group["Price"].map("${:.2f}".format)

purchase_summary = purchase_group[["Purchase Count", "Average Purcahse Price", "Price"
                                   , "Average Purcahse Total per Person"]]
summary_df = purchase_summary.rename(columns={"Price": "Total Purcahse Value"})
summary_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spenders_info = purchase_data[["SN", "Price"]]
spenders_group = spenders_info.groupby("SN").sum()
spenders_group["Purchase Count"] = spenders_info.groupby("SN").count()
spenders_group["Average Purchase Price"] = (spenders_group["Price"])/(spenders_group["Purchase Count"])
organized_group = spenders_group.rename(columns={"Price":"Total Purchase Value"})
top_spenders = organized_group.sort_values("Total Purchase Value", ascending=False)
top_spenders["Total Purchase Value"] = top_spenders["Total Purchase Value"].map("${:.2f}".format)
top_spenders["Average Purchase Price"] = top_spenders["Average Purchase Price"].map("${:.2f}".format)
top_spenders.head()
```

## Most Popular Items

* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_data = purchase_data [["Item ID", "Item Name", "Price"]]
item_group = item_data.groupby(["Item ID", "Item Name"]).sum()
item_group["Purchase Count"] = item_data.groupby(["Item ID", "Item Name"]).count()
item_group["Item Price"] = (item_group["Price"]/item_group["Purchase Count"]).map("${:.2f}".format)
item_group["Price"] = item_group["Price"]
#.map("${:.2f}".format)
organized_item = item_group.rename(columns={"Price":"Total Purchase Value"})
top_items = organized_item.sort_values("Purchase Count", ascending=False)
top_items.head()
```
